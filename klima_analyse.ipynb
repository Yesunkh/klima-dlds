## 1. Data description

DWD Climate Data Center area-mean series (“Gebietsmittel”): annual mean air temperature for Germany and its federal states, 1881–2025, derived from the homogenized station network. One semicolon-separated file (145 rows × 18 columns after cleaning), unit °C, no missing values. Source: opendata.dwd.de (CDC, regional_averages_DE), retrieved 2026-06-15.

## 2. Loading and cleaning

Produces `data/clean/temp_annual_regions.csv` — the shared starting file for the team. Quirks handled: duplicate `Jahr` column, leading spaces, ghost column from the trailing semicolon.

```python
import urllib.request
from pathlib import Path
import pandas as pd

URL = ("https://opendata.dwd.de/climate_environment/CDC/"
       "regional_averages_DE/annual/air_temperature_mean/"
       "regional_averages_tm_year.txt")
RAW = Path("data/raw/regional_averages_tm_year.txt")
if not RAW.exists():
    RAW.parent.mkdir(parents=True, exist_ok=True)
    urllib.request.urlretrieve(URL, RAW)

df = pd.read_csv(RAW, sep=";", skiprows=1, skipinitialspace=True)
df = (df.drop(columns=["Jahr.1"])
        .dropna(axis=1, how="all")
        .rename(columns={"Jahr": "jahr"}))

assert df.shape == (145, 18), df.shape
assert df.isna().sum().sum() == 0
print(df.shape, int(df["jahr"].min()), int(df["jahr"].max()), round(df["Deutschland"].mean(), 2))

df.to_csv("data/clean/temp_annual_regions.csv", index=False)
```
